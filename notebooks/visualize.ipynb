```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
current_file_dir = os.path.dirname(os.path.abspath(__file__)) if '__file__' in globals() else os.getcwd()
BASE_DIR = os.path.abspath(os.path.join(current_file_dir, ".."))

DATA_DIR = os.path.join(BASE_DIR, "data")
MODEL_DIR = os.path.join(BASE_DIR, "models")

file_path = os.path.join(DATA_DIR, 'demos.hdf5')
```

```python
if not os.path.exists(file_path):
    print(f"Warning: No data file found at: {file_path}")
```

```python
if os.path.exists(file_path):
    with h5py.File(file_path, 'r') as f:
        demo_count = len(f) // 2
        states = np.concatenate([f[f'states_{i}'][:] for i in range(demo_count)])
else:
    print(f"File not found: {file_path}")
```

```python
# Plot Joint 1 Position over Time
plt.figure(figsize=(10, 4))
plt.plot(states[:, 0])
plt.title("Joint 1 Position Across Demonstrations")
plt.xlabel("Timestep")
plt.ylabel("Joint Position Value")
plt.grid(True)
plt.show()
```
